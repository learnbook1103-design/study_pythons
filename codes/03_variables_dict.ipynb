문제 3. 딕셔너리 값 바꾸기  
다음과 같은 딕셔너리가 있습니다.  
`"age"` 값을 `"30"`으로 바꾸고, `"name"` 값을 `"Charlie"`로 바꾸는 코드를 작성해 보세요.

```python
info = {
    "name": "Bob",
    "age": "25",
    "city": "Seoul"
}
# "age" 값을 "30"으로 변경
# "name" 값을 "Charlie"로 변경
```

````python
age = '30'
name = 'Charlie'
info = {
    "name": "Bob",
    "age": "25",
    "city": "Seoul"
}
info["age"] = age
info["name"] = name
````

````python
info
````
